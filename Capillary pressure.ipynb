```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


#Creating variable of laboratory capillary pressure

#Interfacial Tension, dynes/cm
sigma = 50 

#Lab Porosity
phi_lab = 0.16

#Lab Permeability
k_lab = 80

#Creating variable of reservoir capillary pressure

#Described porosity by the Reservoir analysis (Actual)
phi = 0.19 #porosity

#Described absolute permeability by the Reservoir analysis (Actual)
k = 120 #md

pc_lab = {'Sw':[1.0,0.8,0.6,0.4,0.25,0.23,0.21,0.20],
      'Pc_lab':[0.5,0.6,0.75,1.05,1.50,1.75,3.0,8.0]}

df = pd.DataFrame(pc_lab)
df

plt.style.use("classic")
plt.figure(figsize=(9,6))
plt.plot(df['Sw'],df['Pc_lab'], label = 'Pc v/s Sw', color = 'red', lw =3)


plt.xlabel('Sw', size = 12)
plt.ylabel('Pc', size = 12)
plt.title("Laboratory Data", size = 20, pad = 22)


plt.legend()
plt.tight_layout()
plt.grid()
plt.show()



df['J'] = 0.21645*(df['Pc_lab']/sigma)*(np.sqrt(k_lab/phi_lab))
df
df['Pc_reservoir'] = (df['J']*sigma)/(0.21645*np.sqrt(k/phi))
df   

plt.style.use("classic")
plt.figure(figsize=(12,6))

plt.plot(df['Sw'],df['Pc_lab'],label = 'Pc(Lab) v/s Sw', lw = 2)
plt.plot(df['Sw'],df['Pc_reservoir'],label = 'Pc(Actual) v/s Sw', lw = 2)

plt.xlabel('Sw(Fraction)', size = 12)
plt.ylabel('Pc(Psi)', size = 12)
plt.title('Application of Leverett-J function', size = 20 ,pad = 22)


plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```
