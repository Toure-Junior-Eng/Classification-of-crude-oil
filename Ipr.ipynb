```python
#create an inflow performance relashioship application
# PI = Q/(Res - Pwf) , Res= reservoir pressure, Pwf= flowing bottom hole pressure

import matplotlib.pyplot as plt

# Step1 : Create Well Test Variables
press = 3500 # psig
pwf = 2800 # psig
rate = 5000 #stb  

# step2 : IPR component calculation 
pi = rate/(press - pwf)
print("Productivity index =", pi, "stb/d/psi")

# Vogel
p_ratio = pwf/press
qomax = rate / (1-0.2*p_ratio - 0.8 *p_ratio**2)
# Generating a list of pwf to get Q...
pwf_list = list( range(0,press+1) )
#Step 3 : Iterate over the pressure list
pi_q_list = []
vogel_q_list = []
for p in pwf_list:
    q_pi = pi * (press - p)
    p_ratio = p/press 
    q_vogel = qomax * (1-0.2*p_ratio - 0.8 *p_ratio**2)
    pi_q_list.append(q_pi)
    vogel_q_list.append(q_vogel)

#Step 4 : plot the inflow curves..
plt.plot(pi_q_list, pwf_list, label="PI")
plt.plot(vogel_q_list, pwf_list, label="Vogel")
#plot setup
plt.ylim(0);plt.xlim(0);plt.legend()
plt.grid();plt.title("Plotting")
```
